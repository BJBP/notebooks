```python
from moviepy.editor import ImageSequenceClip
import os

directorio_fotos = 'Fotos'
fotos = [os.path.join(directorio_fotos, foto) for foto in os.listdir(directorio_fotos) if foto.endswith(('.png', '.jpg', '.jpeg'))]
fotos.sort()  # Opcional: ordenar los archivos si es necesario

clip = ImageSequenceClip(fotos, fps=1/3)

clip.write_videofile('video_final.mp4', codec='libx264')
```
