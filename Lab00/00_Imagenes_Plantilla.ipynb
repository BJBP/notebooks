```python

# Leer la imagen en BGR y escala de grises
x = "../imagenes/_lena.png"
imagen = cv2.imread(x)
imagen2 = cv2.imread(x, 0)

# Convertir la imagen de escala de grises a BGR para poder concatenar con las otras
imagen_gris_bgr = cv2.cvtColor(imagen2, cv2.COLOR_GRAY2BGR)

# Convertir la imagen BGR a RGB
imagen_rgb = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)

# Redimensionar las im√°genes al 50% de su tama√±o original
ancho = int(imagen.shape[1] * 0.5)
alto = int(imagen.shape[0] * 0.5)
dimensiones = (ancho, alto)

imagen_bgr_pequena = cv2.resize(imagen, dimensiones)
imagen_gris_bgr_pequena = cv2.resize(imagen_gris_bgr, dimensiones)
imagen_rgb_pequena = cv2.resize(imagen_rgb, dimensiones)

# Concatenar las im√°genes en una sola ventana
imagen_concatenada = cv2.hconcat([imagen_bgr_pequena, imagen_gris_bgr_pequena, imagen_rgb_pequena])
#imagen_concatenada = cv2.resize(imagen_concatenada, dimensiones)
# Agregar texto a cada imagen
cv2.putText(imagen_concatenada, 'BGR', (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)
cv2.putText(imagen_concatenada, 'Escala de Grises', (ancho + 10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)
cv2.putText(imagen_concatenada, 'RGB', (ancho * 2 + 10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)

# Mostrar la imagen concatenada
cv2.imshow('Imagenes Concatenadas', imagen_concatenada)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import matplotlib.pyplot as plt

# Cargar la imagen
x = "../imagenes/_lena.png"
imagen = cv2.imread(x)

# Convertir la imagen a grises
imagen2 = cv2.imread(x, 0)

# Convertir la imagen grises a BGR
imagen3 = cv2.cvtColor(imagen2, cv2.COLOR_GRAY2BGR)

# Convertir la imagen BGR a RGB
imagen4 = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)

# Crear una figura con 3x1 subplots y apagar los ejes
fig, axs = plt.subplots(1, 4, figsize=(15, 5))

# Iterar a trav√©s de la lista de im√°genes para mostrar
for i, img in enumerate([imagen2, imagen, imagen3, imagen4]):
    # Mostrar la imagen en el subplot correspondiente
    axs[i].imshow(img)
    # Agregar texto en la parte superior de la imagen
    axs[i].text(0.5, 0.95, f'Imagen {i+1}', ha='center', va='center', transform=axs[i].transAxes, fontsize=12, fontweight=0, color='white')
    # Apagar los ejes para el subplot
    axs[i].axis('off')

# Agregar un t√≠tulo a la ventana
plt.suptitle('Imagen en RGB, BGR y escala de grises', fontsize=16, fontweight=0)

plt.show()
```

### 2.5. Lectura de archivo de video

```python
def rescale_frame(frame, op=0, ratio=0.75,W=200,H=100):
    if op==0:
        width = int(frame.shape[1] * ratio)
        height = int(frame.shape[0] * ratio)
        resized_frame= cv2.resize(frame, (width, height), interpolation=cv2.INTER_AREA)
    else:
        resized_frame = cv2.resize(frame, (W, H), interpolation=cv2.INTER_LINEAR)
    return resized_frame
```

```python
# Inicializar lectura de archivo de video
vreader = cv2.VideoCapture('../imagenes/Isles of Glencoe.mp4')

#Mostrar el video frame by frame
while True:
    ret, frame = vreader.read()
    if not ret:
        break
    frame = rescale_frame(frame, ratio=0.75)
    cv2.imshow('Mi video', frame) # Mostrar el frame

    # Salimos presionando: x
    if cv2.waitKey(1) & 0xFF == ord('x'):
        break

# Liberar el lector de video
vreader.release()
cv2.destroyAllWindows()
```

## 3. Crear y grabar

### 3.1. Imagen en escala de grises

```python
d=(80,80)
img=np.zeros(d)
for i in range(20,61):
    for j in range(20,61):
        img[i,j]=255-i-j
img[30:51,30:51]=0;
img=np.uint8(img)
plt.figure(figsize=(3,3))
plt.imshow(img,cmap=plt.cm.gray)
plt.axis("off")
```

### 3.2. Imagen a color

```python
f=c=80
d=(f,c,3)
img=np.zeros(d,dtype=np.uint8)
for i in range(20,61):
    for j in range(20,61):
        img[i,j,0]=img[i,j,1]=0
        img[i,j,2]=255
img[30:51,30:51,0:3]=[255,0,0];
plt.figure(figsize=(3,3))
plt.imshow(img)
plt.axis("off")
```

### 3.4. Grabar video

```python
# Inicializar captura de video en tiempo real solo si tienes una camara instalada
camara = 0   # La camara por defecto es cero
vreader = cv2.VideoCapture(camara)

# Crear VideoWriter:
output = 'captura.mp4'
fps = 30  # Frames x segundo
size = (640, 480)  # Frame size
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec para formato MP4
vwriter = cv2.VideoWriter(output, fourcc, fps, size)

while True:
    ret, frame = vreader.read()
    if not ret:
        break
    cv2.imshow('Captura en tiempo real', frame) # Mostrar el frame
    vwriter.write(frame) # Grabamos el frame
    # Salimos presionando: x
    if cv2.waitKey(1) & 0xFF == ord('x'):
        break
# Liberar el lector de video
vreader.release()
vwriter.release()
cv2.destroyAllWindows()
```
